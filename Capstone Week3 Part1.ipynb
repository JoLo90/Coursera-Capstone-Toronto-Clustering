# Part 1: Prepare the Data

#### FIRST IMPORT THE DATA FROM THE WIKIPEDIA PAGE

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df_raw=pd.read_html(url, header=0)[0]
df_raw.head()
```

#### FILTER THE NOT ASSIGNED BOROUGH

```python
df = df_raw[df_raw['Borough']!='Not assigned']
df.groupby('Postal Code').count().sort_values('Neighborhood', ascending=False)
```
